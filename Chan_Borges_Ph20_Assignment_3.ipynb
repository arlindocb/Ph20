```python
import math
import pprint
import numpy as np
import matplotlib.pyplot as plt 
```

Question 1

```python
h = 0.1
tlst = np.arange(0, 30.1, h)

# Setting initial conditions
xlst = [0]
vlst = [1]

for i in range(len(tlst) - 1):
    xlst.append(xlst[i] + h * vlst[i])
    vlst.append(vlst[i] - h * xlst[i])
    
plt.plot(xlst, vlst, '-')
plt.ylabel('v');
plt.xlabel('x');
plt.show()
```

Question 2

```python
x_anal = np.sin(tlst)
v_anal = np.cos(tlst)

err_x = [(x1 - x2) for (x1, x2) in zip(x_anal, xlst)]
err_v = [(x1 - x2) for (x1, x2) in zip(v_anal, vlst)]

plt.plot(tlst, err_x, '-r', label='x error')
plt.plot(tlst, err_v, '-b', label='v error')
plt.ylabel('error');
plt.xlabel('t');
plt.legend(loc=0)
plt.show()
```

Question 3

```python
h = 0.01
hlst = [h, h/2, h/4, h/8, h/16]
trunc = []

for i in range(len(hlst)):
    xlst1 = [0]
    vlst1 = [1]
    tlst1 = np.arange(0, 30 + 10**-10, hlst[i])
    for j in range(len(tlst1) - 1):
        xlst1.append(xlst1[j] + hlst[i] * vlst1[j])
        vlst1.append(vlst1[j] - hlst[i] * xlst1[j])
    x_anal1 = np.sin(tlst1)
    err_x1 = [(x1 - x2) for (x1, x2) in zip(x_anal1, xlst1)]
    trunc.append(np.max(err_x1))
    
plt.plot(hlst, trunc, 'o-b')
plt.ylabel('max x error');
plt.xlabel('h');
plt.show()
```

Question 4

```python
Elst = [(x**2 + v**2) for (x, v) in zip(xlst, vlst)]

plt.plot(tlst, Elst, '-b')
plt.ylabel('E');
plt.xlabel('t');
plt.show()
```

Question 5

```python
h = 0.1
tlst2 = np.arange(0, 30.1, h)

# Setting initial conditions
xlst2 = [0]
vlst2 = [1]

for i in range(len(tlst2) - 1):
    xlst2.append((xlst2[i] + h * vlst2[i]) / (h**2 + 1))
    vlst2.append(vlst2[i] - h * xlst2[i + 1])
    
err_x2 = [(x1 - x2) for (x1, x2) in zip(x_anal, xlst2)]
err_v2 = [(x1 - x2) for (x1, x2) in zip(v_anal, vlst2)]

plt.plot(tlst2, err_x2, '-r', label='x error')
plt.plot(tlst2, err_v2, '-b', label='v error')
plt.ylabel('error');
plt.xlabel('t');
plt.legend(loc=0)
plt.show()

Elst2 = [(x**2 + v**2) for (x, v) in zip(xlst2, vlst2)]

plt.plot(tlst2, Elst2, '-b')
plt.ylabel('max x error');
plt.xlabel('h');
plt.show()
```
